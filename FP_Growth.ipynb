# ALGORITMO FP GROWTH

#Descripción:

FP-Growth (Frequent Pattern Growth) es un algoritmo utilizado en minería de datos y análisis de patrones en conjuntos de datos. Fue desarrollado para abordar el problema de la extracción de patrones frecuentes en datos transaccionales o conjuntos de elementos, como cestas de compras, registros de transacciones, historiales de navegación web, etc.

El objetivo principal de FP-Growth es encontrar patrones frecuentes en un conjunto de datos de manera eficiente, sin la necesidad de generar explícitamente todos los subconjuntos posibles de elementos. Esto hace que el algoritmo sea mucho más rápido y escalable en comparación con otros enfoques tradicionales como el algoritmo Apriori.

El proceso general del algoritmo FP-Growth es el siguiente:

* Construcción del árbol FP-Tree: Primero, se crea una estructura de datos llamada FP-Tree (Frequent Pattern Tree) a partir de los datos de entrada. Cada ruta desde la raíz hasta una hoja en el árbol representa un patrón. Los nodos del árbol almacenan información sobre los elementos y su frecuencia en los datos.

* Construcción de tablas de enlace: Se crea una tabla de enlace para cada elemento en el conjunto de datos, que conecta los elementos con los nodos correspondientes en el FP-Tree.

* Extracción de patrones frecuentes: A partir del FP-Tree y las tablas de enlace, el algoritmo extrae los patrones frecuentes al realizar un recorrido ascendente del árbol. Esto evita la generación explícita de subconjuntos y permite identificar patrones con mayor eficiencia.

* Generación de conjuntos de patrones condicionales: El proceso se repite recursivamente para generar conjuntos de patrones condicionales, lo que ayuda a construir árboles FP más pequeños y a realizar el proceso de extracción de patrones con mayor rapidez.


# Donde puede ser implementado #
* Reglas de Asociación: El algoritmo FP-Growth es especialmente útil para la extracción de reglas de asociación en conjuntos de datos transaccionales. Puede ayudar a identificar patrones frecuentes en la compra de productos, navegación en sitios web, selección de servicios, entre otros.

* Cesta de Compras: En análisis de cestas de compra, el algoritmo FP-Growth puede revelar patrones frecuentes de productos que los clientes tienden a comprar juntos. Esto es valioso para estrategias de marketing, recomendación de productos y optimización de disposición de tiendas.

* Secuencias de Eventos: Aunque no es la aplicación principal del FP-Growth, también puede ser utilizado en secuencias de eventos discretos para identificar patrones frecuentes, como acciones en videojuegos, interacciones en redes sociales o secuencias de clics en un sitio web.

* Detección de Fraudes: En sistemas de detección de fraudes, el FP-Growth puede encontrar patrones frecuentes de comportamientos sospechosos que puedan indicar actividades fraudulentas. Por ejemplo, patrones de transacciones inusuales o comportamientos anómalos.

* Análisis de Logs: En sistemas de registro y monitoreo, el FP-Growth puede ser utilizado para identificar patrones frecuentes en registros de eventos, como actividades de usuarios, fallas de sistemas o patrones de acceso.

* Biología y Genética: En bioinformática, el FP-Growth puede ayudar a identificar patrones frecuentes en secuencias de ADN o proteínas, lo que podría conducir al descubrimiento de relaciones significativas en genómica y biología molecular.

* Market Basket Analysis: La técnica del análisis de cesta de compra, que busca patrones frecuentes en los artículos que los clientes compran juntos, es un caso clásico de aplicación del algoritmo FP-Growth.

* Optimización de Inventario: En la gestión de inventario y suministros, el FP-Growth puede identificar patrones frecuentes de productos solicitados en un mismo pedido, lo que ayuda en la planificación de inventario y la eficiencia logística.

#Supuestos y Restricciones



#Supuestos:

* Conjunto de Datos Transaccional: El algoritmo FP-Growth está diseñado principalmente para conjuntos de datos transaccionales, donde cada transacción es una lista de elementos. No es adecuado para todos los tipos de datos.

* Datos Discretos: El algoritmo FP-Growth trabaja mejor con datos discretos, como elementos de una lista. No es adecuado para datos continuos o valores numéricos sin discretizar.

# Restricciones:

* Uso de Memoria: El árbol FP puede requerir una cantidad significativa de memoria, especialmente para conjuntos de datos grandes con muchos elementos únicos. Esto puede limitar su aplicabilidad en sistemas con recursos limitados.

* Orden de los Elementos: El rendimiento del algoritmo FP-Growth puede depender del orden en que se presentan los elementos en las transacciones. Por lo general, se requiere ordenar los elementos según su frecuencia antes de aplicar el algoritmo para obtener mejores resultados.

* Ajuste del Parámetro de Soporte: La eficacia del algoritmo FP-Growth puede depender de la elección del umbral de soporte mínimo. Si el valor es demasiado bajo, puede llevar a la generación de un gran número de patrones, mientras que un valor muy alto puede hacer que se pierdan patrones interesantes.

* Extracción de Reglas de Asociación: Aunque el algoritmo FP-Growth es eficiente para encontrar patrones frecuentes, la generación de reglas de asociación a partir de estos patrones puede requerir recursos adicionales y no es inherentemente parte del algoritmo FP-Growth en sí.

* No Maneja Datos Temporales o Secuenciales: El algoritmo FP-Growth no tiene en cuenta el orden temporal o secuencial de las transacciones. Si se necesitan patrones frecuentes en datos temporales o secuenciales, se requiere un enfoque diferente.


#Ejemplo en codigo

```python
# import mlxtend.frequent_patterns as fpgrowth
from mlxtend.frequent_patterns import fpgrowth
import pandas as pd

dataset = pd.read_csv('Groceries_dataset.csv')
dataset

```

```python

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

split_items = dataset['itemDescription'].str.split(',')


```

```python

encoded_dataset = pd.get_dummies(split_items.apply(pd.Series).stack()).sum(level=0)
encoded_dataset

fpgrowth(encoded_dataset, min_support = 0.01, use_colnames = True)
```

```python
frequent_itemsets = fpgrowth(encoded_dataset, min_support = 0.01, use_colnames = True)
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

# Add the root node
G.add_node("Root")

for itemset in frequent_itemsets:
    parent = "Root"
    for item in itemset:
        if not G.has_node(item):
            G.add_node(item)
        G.add_edge(parent, item)
        parent = item

# Plot the tree-like structure
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=3000, node_color='lightblue')
plt.show()
```
